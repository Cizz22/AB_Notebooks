## Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Read CSV

```python
data = pd.read_csv('/Users/user/AB/data/Sleep_Efficiency.csv')
```

## Line graph using matplotlib

```python
data
```

```python
sns.lmplot(x='Age', y='Sleep efficiency', data=data, hue="Gender")
plt.title('Scatter Plot Sleep Efficiency vs Age')
```

```python

# Set theme
sns.set_style('whitegrid')
 
# Violin plot
plt.figure(figsize=(10,8)) # Set plot dimensions
sns.violinplot(x='Smoking status', y='Sleep efficiency', data=data)

plt.title('Violin plot of Sleep Efficiency vs Smoking Status')
```

```python
# Categorical Plot
g = sns.catplot(x='Exercise frequency',
                   y='Sleep efficiency', 
                   data=data, 
                   hue='Smoking status',  # Color by stage
                   col='Smoking status',  # Separate by stage
                   kind='swarm') # Swarmplot
 
# Rotate x-axis labels
g.set_xticklabels(rotation=-45)

plt.title('Categorical Plot of Sleep Efficiency vs Exercise Frequency and Smoking Status')
```
